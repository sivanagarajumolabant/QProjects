```python
from langgraph.graph import Graph
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()
```

```python
def chatbot(input):
    llm = ChatGroq(model="llama3-70b-8192")
    # llm.invoke('Hi').content
    response = llm.invoke(input).content
    return response


def upper_data(data):
    return data.upper()
```

```python
flow = Graph()
flow.add_node("llm_data", chatbot)
flow.add_node("upper_data", upper_data)
flow.add_edge("llm_data", "upper_data")
# flow.add_edge('upper_data','llm_data')
```

```python
flow.set_entry_point("llm_data")
flow.set_finish_point("upper_data")
app = flow.compile()  #
```

```python
from IPython.display import Image, display

try:
    display(Image(app.get_graph().draw_mermaid_png()))
except Exception as e:
    print(e)
```
